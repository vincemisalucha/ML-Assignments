```python
import pandas as pd
import datetime
```

```python
data = pd.DataFrame(data)
```

```python
#Condition 1: The preceeding day in reference to the first missing date should not be empty
def StockDataImputation(data, missing_dates):
    
    for i in range(missing_dates.shape[0]):
        
        if missing_dates['Day'][i] == 0:
        
            datetime_date = missing_dates[0][i].to_pydatetime()
            datetime_date = datetime_date.date() - datetime.timedelta(3)
            datetime_date

            data.loc[pd.Timestamp(missing_dates[0][i])] = [data.loc[pd.Timestamp(datetime_date), 'Open'],
                                                   data.loc[pd.Timestamp(datetime_date), 'High'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Low'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Close'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Adj Close'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Volume']]

            data = data.sort_values(by='Date', ascending=True)
            
        else:
            
            datetime_date = missing_dates[0][i].to_pydatetime()
            datetime_date = datetime_date.date() - datetime.timedelta(1)
            datetime_date

            data.loc[pd.Timestamp(missing_dates[0][i])] = [data.loc[pd.Timestamp(datetime_date), 'Open'],
                                                   data.loc[pd.Timestamp(datetime_date), 'High'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Low'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Close'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Adj Close'],
                                                   data.loc[pd.Timestamp(datetime_date), 'Volume']]

            data = data.sort_values(by='Date', ascending=True)
            
    return data
```

```python
missing_values = pd.date_range(start='2003-12-31', end='2021-08-31').difference(data.index)
```

```python
type(missing_values)
```

```python
days = []
for x in missing_values:
    days.append(x.weekday())
```

```python
final_miss = pd.DataFrame(missing_values)
final_miss['Day'] = days
final_miss.head()
```

```python
final_miss = final_miss.loc[final_miss['Day'] != 5]
final_miss = final_miss.loc[final_miss['Day'] != 6]
final_miss
```

```python
final_miss = final_miss.reset_index(drop=True)
final_miss
```

```python
final_miss.shape[0]
```

```python
imputed_data = StockDataImputation(data, final_miss)
```
