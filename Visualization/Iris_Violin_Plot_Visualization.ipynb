```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
```

```python
data.describe()
```

```python
sns.heatmap(data.iloc[:, :-1].corr(), annot=True)
```

```python
df = pd.read_csv('C:/Users/Admin/Desktop/iris data.txt', delimiter = ",", 
                 names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"])
```

```python
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,10))

color = {"Iris-versicolor": "orange", "Iris-setosa": "blue", "Iris-virginica": "red"}
          
sns.violinplot( y="petal_width", x= "species", data=df, orient='v' , ax=axes[0, 0],inner='quartile', palette=color)
sns.violinplot( y="petal_length", x= "species", data=df, orient='v' , ax=axes[0, 1],inner='quartile',  palette=color)
sns.violinplot( y="sepal_length", x= "species", data=df, orient='v' , ax=axes[1, 0],inner='quartile',  palette=color)
sns.violinplot( y="sepal_width", x= "species", data=df, orient='v' , ax=axes[1, 1],inner='quartile',  palette=color)
plt.show()
```
